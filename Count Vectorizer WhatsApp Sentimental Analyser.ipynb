```python
 def sentimentalAnalysis(data,columnname):
        
        #downloading vader_lexicon for the process
        import matplotlib.pyplot as plt
        import seaborn as sns
        "Importing Necessary Packeage"
        from nltk.sentiment.vader import SentimentIntensityAnalyzer
        sid = SentimentIntensityAnalyzer()
        
        "Deleting null pr empty value"
        data.dropna(inplace=True)
        
        "Checking for a comment"
        #sid.polarity_scores(data[columnname][93])
        
        "Creating respective columns"
        
        data['scores'] = data[columnname].apply(lambda commentText: sid.polarity_scores(commentText))
        data['compound']  = data['scores'].apply(lambda score_dict: score_dict['compound'])
        data['Negtive']  = data['scores'].apply(lambda score_dict: score_dict['neg'])
        data['Postive']  = data['scores'].apply(lambda score_dict: score_dict['pos'])
        data['Neutral']  = data['scores'].apply(lambda score_dict: score_dict['neu'])
        
        "Creating final pos or neg using compound score"
        data['comp_score'] = data['compound'].apply(lambda c: 'pos' if c >=0 else 'neg')
        plt.clf()
        #comp=sns.countplot(x = 'comp_score', hue = 'Name', data = data, palette = 'magma')
        #comp.figure.savefig("date_charts.png")
        "Checking how many pos and neg"
        posneg=pd.DataFrame(data['comp_score'].value_counts())
        return posneg,data
    
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
CV=CountVectorizer(max_df=0.95,min_df=2,stop_words='english')
dtm=CV.fit_transform(df["Chat"])
```

```python
from sklearn.decomposition import NMF
nmf_model=NMF(n_components=5,random_state=42)
nmf_model.fit(dtm)
```

```python
for index,topic in enumerate(nmf_model.components_):
    results=([CV.get_feature_names_out()[i] for i in topic.argsort()[-10:]])
    print(results)
```

```python
topic_results=nmf_model.transform(dtm)
df["Topic"]=topic_results.argmax(axis=1)
```

```python
# Convert 'Date' column to datetime format
df['Date1'] = pd.to_datetime(df['Date'],format='%d/%m/%y')
#print(df['Date'])

# Extract most active date
most_active_date = df['Date1'].value_counts().idxmax()
most_active_date_str = most_active_date.strftime('%Y-%m-%d')

# Extract most active day
most_active_day = most_active_date.strftime('%A')

# Extract most active time
most_active_time = df['Time'].value_counts().idxmax()


print("Most active date:", most_active_date_str)
print("Most active day:", most_active_day)
print("Most active time:", most_active_time)
```

```python
df['num_words'] = df['Chat'].apply(lambda x: len(nltk.word_tokenize(x)))

# Calculate average number of words per message for each name
average_words_per_message = df.groupby('Name')['num_words'].mean()

# Determine the threshold for talkative vs less talkative
median_num_words = average_words_per_message.median()

# Classify each name as Talkative or Less talkative
def classify_talkativeness(avg_words):
    if avg_words >= median_num_words:
        return 'Talkative'
    else:
        return 'Less talkative'

df['talkativeness'] = df['Name'].map(average_words_per_message.apply(classify_talkativeness))

# Drop duplicates to ensure each name appears only once in the result
unique_names = df.drop_duplicates(subset='Name')

# Iterate over the unique names and print the result
for index, row in unique_names.iterrows():
    print(f"{row['talkativeness']}: {row['Name']}")
```

```python
import pandas as pd
import re

# Define regular expressions for different types of content
media_regex = r'\[image\]|\[video\]'
deleted_message_regex = r'Did you see my deleted message\?'
missed_voice_call_regex = r'Missed voice call from (\w+)\.'
missed_video_call_regex = r'Missed video call from (\w+)\.'


# Count occurrences of each type of content
total_media_count = df['Chat'].str.count(media_regex).sum()
deleted_message_count = df['Chat'].str.count(deleted_message_regex).sum()
missed_voice_call_count = len(re.findall(missed_voice_call_regex, ' '.join(df['Chat'])))
missed_video_call_count = len(re.findall(missed_video_call_regex, ' '.join(df['Chat'])))

print("Total media count:", total_media_count)
print("Deleted message count:", deleted_message_count)
print("Missed voice call count:", missed_voice_call_count)
print("Missed video call count:", missed_video_call_count)
```

```python
import pandas as pd


# Define keywords indicative of flirtatious behavior
flirt_keywords = ['beautiful', 'sweet', 'thinking about you', 'dinner', 'date','cute','love','hot','sexy','hug']

# Function to calculate flirt percentage for each name
def calculate_flirt_percentage(name):
    # Filter chat messages for the specific name
    name_messages = df[df['Name'] == name]['Chat']
    total_messages = len(name_messages)
    
    # Count flirtatious messages
    flirt_messages = name_messages.apply(lambda x: any(keyword in x.lower() for keyword in flirt_keywords)).sum()
    
    # Calculate flirt percentage
    flirt_percentage = (flirt_messages / total_messages) * 100 if total_messages > 0 else 0
    return flirt_percentage

# Calculate flirt percentage for each unique name
unique_names = df['Name'].unique()
flirt_percentages = {name: calculate_flirt_percentage(name) for name in unique_names}

print("Flirt percentages:")
for name, flirt_percentage in flirt_percentages.items():
    print(f"{name}: {flirt_percentage:.2f}%")
```
